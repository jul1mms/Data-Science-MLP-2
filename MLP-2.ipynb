```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("weather_data.csv")
df.head()
```

Gráfico de Linha: Mostra a variação da temperatura ao longo do tempo (índice dos dados).

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Temperature_C'].values, marker='o', linestyle='-', markersize=2)
plt.title('Temperatura ao Longo do Tempo')
plt.xlabel('Índice')
plt.ylabel('Temperatura (C)')
plt.show()
```

Gráfico de Barra: Apresenta a distribuição de frequências das temperaturas (arredondadas).

```python
temp_counts = df['Temperature_C'].round().value_counts().sort_index()

#gráfico
plt.figure(figsize=(10, 6))
temp_counts.plot(kind='bar')
plt.title('Distribuição das Temperaturas')
plt.xlabel('Temperatura (C)')
plt.ylabel('Frequência')
plt.show()
```

Histograma: Fornece uma visualização da distribuição das temperaturas em intervalos (bins).

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Temperature_C'], bins=20, edgecolor='black')
plt.title('Histograma das Temperaturas')
plt.xlabel('Temperatura (C)')
plt.ylabel('Frequência')
plt.show()
```

Gráfico de Pizza: Mostra a proporção de diferentes faixas de temperatura.

```python
temp_bins = pd.cut(df['Temperature_C'], bins=[-20, -10, 0, 10, 20, 30, 40])
temp_bin_counts = temp_bins.value_counts()

#gráfico
plt.figure(figsize=(8, 8))
temp_bin_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribuição das Temperaturas')
plt.ylabel('')
plt.show()
```

Gráfico Scatter: Analisa a relação entre a temperatura e a umidade.

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Temperature_C'], df['Humidity_pct'], alpha=0.5)
plt.title('Temperatura vs Umidade')
plt.xlabel('Temperatura (C)')
plt.ylabel('Umidade')
plt.show()
```
